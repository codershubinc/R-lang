```python
import matplotlib.pyplot as plt


# Data for the plot
x = []
y = []

for i in range(100):
    x.append(i)
    y.append(i**2)


plt.plot(x, y)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], linestyle='--', color='r', marker='o', )
plt.ylabel('some numbers')
plt.xlabel('some numbers')
plt.title('My first plot')


plt.show()
```

```python
labels = x[-10:]
sizes = y[-10:]

plt.figure(figsize=(8, 8))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    explode=(0, 0, 0, 0, 0, 0, 0, 0, 0, .4),
    # hatch='',
    textprops={'fontsize': 14},
    radius=1.5,
    shadow=True,

)
plt.title('Pie Chart of Last 10 Elements')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
dict = {
    'red': 10,
    'blue': 20,
    'green': 30,
    'yellow': 40,
    'orange': 50,
    'purple': 60,
    'black': 70,
    'white': 80,
    'pink': 90,
    'brown': 100
}

plt.bar(
    dict.keys(),
    dict.values(),
    color=dict.keys(),
    edgecolor='black',
)

plt.show()

plt.pie(
    dict.values(),
    labels=dict.keys(),
    colors=dict.keys(),
    autopct='%1.1f%%',

)
plt.show()
```
