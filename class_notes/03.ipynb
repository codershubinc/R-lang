```python
import numpy as np
import pandas as pd
import seaborn as sns


val = np.random.randint(0, 100, 20)
dt = np.reshape(val, (5, 4))

df = pd.DataFrame(
    dt,
    columns=['A', 'B', 'C', 'D']
)

print(df)

sns.heatmap(
    data=df,
    annot=True,
    cmap="crest",
    linewidths=5,
    linecolor='blue',
)
```

```python
ds = sns.load_dataset("tips")
print(ds)
sns.countplot(
    data=ds,
    x='smoker',
    gap=.5,
    palette=["lightblue", "lightgreen"],
    edgecolor='black',
    linewidth=2,
)
```
